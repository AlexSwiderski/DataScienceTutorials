### Data Science
Though a somewhat vague term in today's standards, a data scientist is an individual who aims to understand data. They often, if not always, utlize a specific computer language to visualize, describe, and make inference of data. As you embark on the journey of becoming a data scientist, which perhaps will never end, do not lose focus of your job.  Regardless of the task at hand, be it analyzing massive ammounts of data from google text, analyzing results from a pschological assessament, or investigating spike train data from single firing neurons in the hippocampus, do not lose track of your goal.  Although each data will tell a different story, your goal is always to gain insight what story the data is telling.

## Rule number 1:  While keping an eye on the future, never forget your passed. 

In my experience with data science and data exploration there is ####[always] a next step.  While you exploring the story of your data, never lose track of what you have already done. Data science is MESSY.  You will fail, and fail again. You will fail so many times you won't remember what you failed on last week.  But with every failure comes a solution to a problem.  So while you will always be trying to tackle the next problem, remember to document your passed.  This is why tools such as Jupyter Notebooks, R script files, and R markdown are so helpful.  They let you create your own map, while saving any success you wish to remember.  One very nifty aspect of Jupyter notebooks is the ability for inline text processing, ease of writing complex mathmatical equations with a computer keyboard, and computer code.  The computer coding language we will be working with the   [R Programming Language](https://www.r-project.org/)
