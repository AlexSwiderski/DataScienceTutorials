```r
#create a random distribution of 100 numbers between 0 and 100 under the variable name x
#You'll do this by taking random draws from a uniform distribtion. R does this with the
#function 'runif', which requires that you specify three arguments (pieces of information):
#sample size (n), the minimum value, and the maximum value.
x = runif(n = 100, min = 0, max = 100)
#create data frame named data, with one variable named x that is equal to x above.
data = data.frame(x = x)
```

```r
#create plot with ggplot
#for help with any R function type ?? then the function name
??ggplot
plot1 = ggplot(data = data, (aes(x = x))) +
geom_histogram() +
ggtitle("Figure A")
print(plot1)
```

### Moving on we will look at the normal distribution in more detail.  
Normal distributions are often used to represent real valued random variables for which we may not know the actual distribution but which we plasibly expect to follow a normal distirbution, and the distribution is presented as follows: 
$$
  X \sim \mathcal{N}(\mu,\,\sigma^{2})\,.
$$

Where X is normally distributed as denoted by  $\mathcal{N}$ with a mean, $ \mu $, and variance, $\sigma^{2}$.  Note that the standard deviation is simply the square root of the variance.  Below we are going to simulate a normal distribution, to then learn to estimate mean and standard deviations with R.

Another notable characteristic of the normal dsitribution is that 95% of the distribution falls within $+/-$ two standard deviations from the mean and 68% of the distribution falls within $+/-$ 1 standard deviations of the mean.

```r
#simulate normal distribution named nd within the dataframe named data
data = data.frame(nd = rnorm(1000, mean=0, sd=2))
```

### Let's take a look at what a normal distribution looks like

```r
#create histogram again
#Note the histogramnow has a blue interior 
plot2 = ggplot(data = data, (aes(x = nd))) +
geom_histogram(color="black", fill = "blue") +
ggtitle("Normal Distribution") +
theme(plot.title = element_text(hjust = 0.5))
print(plot2)
```
