### Prior to running any analyses, describe the distributions in terms of variability.  Is there a lot?  A little? The plots are called Histograms.  On the x axis numbers are presented and on the y axis are the amount of times this number was observed.
- For the sake of this excercise

```r
#load libraries
library(ggplot2) #ggplot2 uses for plotting
```

```r
x = runif(10, min = 0, max = 100)
data = data.frame(x = x)
```

```r
plot1 = ggplot(data = data, (aes(x = x))) +
geom_histogram()
print(plot1)
```
